# 图像像素值统计

像素统计：图像直方图、像素最大最小值、均值、标准方差等等。

- 最大值可以用于单通道图像的规划；
- 均值用于二值分割方差，方差为0时就表示图像中像素值没有差异，可能是一张纯色图；
- 最大最小值用于归一化处理，把图像传到CNN进行识别

## `minMaxLoc(src[, mask]) -> minVal, maxVal, minLoc, maxLoc`求取最大最小值

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

src = cv.imread('images/wm.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(src)
minVal, maxVal, minLoc, maxLoc = cv.minMaxLoc(src)
print('min: %.2f, max: %.2f' % (minVal, maxVal))
print('minLoc: ', minLoc)
print('maxLoc: ', maxLoc)
```

## `meanStdDev(src[, mean[, stddev[, mask]]]) -> mean, stddev` 均值标准方差

```python
mean, stdDev = cv.meanStdDev(src)
print('mean: %.2f, stdDev: %.2f' % (mean, stdDev))
```

```python
# 通过均值将图像编程二值图像
src[np.where(src < mean)] = 0
src[np.where(src >= mean)] = 255
plt.imshow(src)




```
