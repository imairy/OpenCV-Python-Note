# 图像像素读写操作

## 通道数
在像素点当中，每一个像素点表示一个颜色，而每个颜色是由RGB三种颜色搭配出来的，通过给RGB不同的值也就行程不同的颜色，这就是通道数的作用。如果图像支持透明，那么就是4通道，除了RGB通道外还有一个透明通道。

当通道数等于3，就是一个彩色图像；通道数为1时，就是一张黑白图像。在图像数据的`shape`属性包含了宽高以及通道数。

```python
import cv2 as cv
import utils

src = cv.imread('images/opencv-logo.png')
utils.imshow(src)
h, w, ch = src.shape
print('h, w, ch : ', h, w, ch)
```

## 读写像素

`src`有3个通道，由于`imread`读取的图像通道顺序是BGR的顺序，所以，0-1-2分别对应蓝色-绿色-红色通道。一项对图像的所有像素点遍历，并将像素点颜色取反，最后将取反后的通道值重新赋值给像素点，也就是说将会得到一张src的反相图。

```python
for row in range(h):
    for col in range(w):
        b ,g ,r = src[row, col]
        b = 255 - b
        g = 255 - g
        r = 255 - r
        src[row, col] = [b ,g, r]
utils.imshow(src)


```
