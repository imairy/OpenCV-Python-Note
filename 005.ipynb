# 像素算术操作

两幅图像如果进行算术预算，类型必须一致，通道数必须一致，大小宽高必须相同。

```python
import cv2 as cv
import numpy as np
import utils as plt

src = cv.imread('images/opencv-logo.png')

m1 = np.zeros(src.shape, src.dtype)
```

## `add(src1, src2[, dst[, mask[, dtype]]]) -> dst`

```python
add_result = np.zeros(src.shape, src.dtype)
cv.add(src, m1, add_result)
plt.imshow(add_result)
```

## `subtract(src1, src2[, dst[, mask[, dtype]]]) -> dst`

```python
sub_result = np.zeros(src.shape, src.dtype)
cv.subtract(src, m1, sub_result)
plt.imshow(sub_result)
```

## `multiply(src1, src2[, dst[, scale[, dtype]]]) -> dst`

```python
mul_result = np.zeros(src.shape, src.dtype)
cv.multiply(src, m1, mul_result)
plt.imshow(mul_result)
```

## `divide(src1, src2[, dst[, scale[, dtype]]]) -> dst`

```python
div_result = np.zeros(src.shape, src.dtype)
cv.divide(src, m1, div_result)
plt.imshow(div_result)

```
