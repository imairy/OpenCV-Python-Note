# 像素操作之逻辑操作

像素的逻辑操作是像素点的位操作。

通过下面代码来看一下逻辑运算的效果

```python
import cv2 as cv
import numpy as np
import utils as plt
```

创建一个400x400的三通道空白图，将空白图的横向50-250和竖向50-250这部分区域的G通道值设为255，R通道值也设为255，就得到一个200x200的黄色矩形（opencv的通道顺序是BGR，但pyplot的通道顺序是RGB，所以预览图为青色）。

```python
m1 = np.zeros([400, 400, 3], dtype=np.uint8)
m1[50:250, 50:250, 1] = 255
m1[50:250, 50:250, 2] = 255
plt.imshow(m1)
```

用同样的方式创建另外一张图，为了让第二张图中的矩形与第一张图产生交错，设置第2张图的高和宽在150-350这个区域向通道值，将B通道设置为255，R通道设置为255。

```python
m2 = np.zeros([400, 400, 3], dtype=np.uint8)
m2[150:350, 150:350, 0] = 255
m2[150:350, 150:350, 2] = 255
plt.imshow(m2)
```

```python
# 与操作
dst_and = cv.bitwise_and(m1, m2)
plt.imshow(dst_and)
```

```python
# 或操作
dst_or = cv.bitwise_or(m1, m2)
plt.imshow(dst_or)

```

```python
# 与或
dst_xor = cv.bitwise_xor(m1, m2)
plt.imshow(dst_xor)
```

```python
# 取反，取反操作常用于对二值图取反
dst_not = cv.bitwise_not(m1)
plt.imshow(dst_not)
```

在与、或、异或的操作方法中最后一个参数是`mask`，`mask`的作用是取指定的区域，可用于取不规则图像。
- `bitwise_and(src1, src2[, dst[, mask]]) -> dst`
- `bitwise_or(src1, src2[, dst[, mask]]) -> dst`
- `bitwise_xor(src1, src2[, dst[, mask]]) -> dst`
