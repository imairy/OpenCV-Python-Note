# 几何形状绘制

OpenCV没有专门的填充方法，只是把绘制几何形状时候的线宽 - `thickness`参数值设置为负数即表示填充该几何形状或者使用参数`CV_FILLED`

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 创建空白图像
src = np.zeros([400, 400, 3], dtype=np.uint8)
plt.imshow(src)
```

## `rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img`矩形
- `pt1`表示绘制矩形左上角相对于`img`的坐标
- `pt1`右下角坐标
- `color`绘制所用的颜色（B, G, R）
- `thickness`线宽
- `lineType`
    - `LINE_8`
    - `LINE_4`
    - `LINE_AA`
- `shift`坐标中的小数位数

```python
# 画矩形
dst = cv.rectangle(src, (100, 150), (200, 200), (255, 0, 255), 2, cv.LINE_4, 0)
plt.imshow(dst)
```

## `circle(img, center, radius, color[, thickness[, lineType[, shift]]]) -> img`圆
- `center`圆心位置
- `radius`半径

```python
dst = cv.circle(src, (250, 250), 100, (255, 255, 0), 2, cv.LINE_8, 0)
plt.imshow(dst)
```

## `ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img`椭圆
- `axes`椭圆的（长轴,短轴)
- `angle`角度
- `startAngle`开始角度
- `endAngle`结束角度

```python
dst = cv.ellipse(src, (200, 200), (50, 30), 90, 0, 270, (0, 255, 255), 2, cv.LINE_8, 0)
plt.imshow(dst)
```

## 填充

```python
dst = cv.rectangle(src, (100, 150), (200, 200), (255, 0, 255), -1, cv.LINE_4, 0)
dst = cv.circle(dst, (250, 250), 100, (255, 255, 0), -1, cv.LINE_8, 0)
dst = cv.ellipse(dst, (200, 200), (50, 30), 90, 0, 270, (0, 255, 255), -1, cv.LINE_8, 0)
plt.imshow(dst)
```

## 随机数
np.random.rand() 表示numpy中随机数生成，生成浮点数0～1的随机数, 包含0，不包含1。

```python
#随机绘制线条

src = np.zeros([400, 400, 3], dtype=np.uint8)

for i in range(10):
    # 创建两个点坐标
    x1 = np.random.rand() * 400
    y1 = np.random.rand() * 400
    x2 = np.random.rand() * 400
    y2 = np.random.rand() * 400

    b = np.random.randint(0, 256)
    g = np.random.randint(0, 256)
    r = np.random.randint(0, 256)

    dst = cv.line(src, (np.int(x1), np.int(y1)), (np.int(x2), np.int(y2)), (b, g, r), 2, cv.LINE_8,0 )
    plt.imshow(dst)
```
