# 图像的创建和赋值

```python
import cv2 as cv
import numpy as np
import utils
from matplotlib import pyplot as plt

src = cv.imread('images/opencv-logo.png')
```

## 图像克隆和拷贝
通过`imread`读取的图像数据是一个`mat`类型（在python中是numpy），使用`copy`方法进行图像的拷贝


```python
m1 = np.copy(src)
utils.imshow(m1)
```

在opencv中通过`copy`的图像是对图像的`mat`进行了一次克隆，也就是说`src`与`m1`指向的是两个不同的内存地址

![数据结构](res/src_copy_to_m1.png)

除了通过`copy`方法对图像进行克隆，还可以使用赋值的方式创建一个新的`mat`对象。赋值方式与`copy`方式的区别在于，通过赋值方式创建的图像对象与原图像对象指向同一个数据块（两者的区别有点像函数中的传值与传引用），下面代码中的的`m2`与`src`指向同一个内存地址。

```python
m2 = src
utils.imshow(m2)
```

在python中可以很简单的给图像的某部分像素点进行修改，比如下面代码，在高度从120到240区域，宽度20到40的区域，这部分的所有像素点设置成255，也就是白色。(预先知道是三通道)

```python
m2[120:240, 20:40, :] = 255
utils.imshow(m2)
```

## `mat`对象数据结构
`mat`对象的基本结构有两部分组成，头部和数据部分。

**头部**包含了一个图的`mat`头信息，比如宽、高、通道数、深度、类型等。

**数据部分**主要包含的图像上像素点的值

## 创建空白图像
除了通过赋值或者`copy`的方式创建一个图像外，还是直接给定数据进行图像的创建，下面代码根据`src`图像的大小和数据类型`dtype`创建一个空白图像：

```python
m3 = np.zeros(src.shape, src.dtype)
utils.imshow(m3)
```

除此之外还可以手动指定所创建图像的大小和数据类型，下面代码创建一个512x512的单通道8-bit图像

```python
m4 = np.zeros([512, 512], dtype=np.uint8)
plt.imshow(m4)
```

创建一个512x512的3通道8-bit图

```python
m5 = np.zeros([512, 512, 3], dtype=np.uint8)
utils.imshow(m5)
```

另外还有通过`np.ones`的方法创建图像，效果同`np.zeros`相同，是从matlib借鉴来的方法。

> 空白对象的应用场景：自定义滤波器、卷积

在opencv中通过`imread`读进来的图像数据的色彩通道是BGR顺序的，那么索引为0的通道就是蓝色通道，1就是绿色通道，2就是红色通道

```python
# 将通道0（蓝色通道）设置为255
m5[:, :, 2] = 255
utils.imshow(m5)



```
