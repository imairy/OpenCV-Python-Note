# 通道分离与合并

opencv通过`imread`读取的图像是一个3通道的彩色图像，使用RGB色彩空间，RGB色彩空间与设备无关，但当使用了一些诸如YUV这种设备相关的色彩空间，便不能在以外的设备上很好的显示图像色彩。

下面将一朵花的图像进行通道分离，剔除绿色和红色通道,仅保留蓝色通道，再将通道合并。

```python
import cv2 as cv
from utils import imshow

src = cv.imread('images/flower.png')
imshow(src)
mv = cv.split(src)
```

```python
# 将绿色通道置为0
mv[1][:,:] = 0
# 将蓝色通道置为0
mv[2][:,:] = 0
dst = cv.merge(mv)
imshow(dst)
```

```python
cv.imshow('image', dst)
cv.waitKey(0)
cv.destroyAllWindows()

```
