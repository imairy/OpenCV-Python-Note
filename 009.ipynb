# 图像色彩空间转换
常见的四种色彩空间：RGB、HSV、YUV、YCrCb：
- **RGB**：设备无关
- **HSV**：直方图相关处理会先转换到HSV色彩空间
- **YUV**：欧洲电视早期保准，另外Linux的相机也是YUV，Android手机的相机也是采用YUV色彩空间，但是会被转换成RGB色彩空间
- **YCrCb**：皮肤检测效果好

## `cvtColor(src, code[, dst[, dstCn]]) -> dst`色彩空间转换
色彩空间间的转换使用`cvtColor`方法：

```python
import cv2 as cv
import utils as u
from matplotlib import pyplot as plt

src = cv.imread('images/flower.png')
hsv_dst = cv.cvtColor(src, cv.COLOR_BGR2HSV)
plt.imshow(hsv_dst)
```

```python
yuv_dst = cv.cvtColor(src, cv.COLOR_BGR2YUV)
plt.imshow(yuv_dst)
```

```python
ycrcb_dst = cv.cvtColor(src, cv.COLOR_BGR2YCrCb)
plt.imshow(ycrcb_dst)
```

## `inRange(src, lowerb, upperb[, dst]) -> dst`

HSV色彩空间的H取值范围是0~180（取决于opencv开发初期内存比较昂贵），S的范围是0~255，V的范围也是0~255。

![](res/hsv.png)

hmin表示H通道的最小值，hmax表示H通道的最大值，SV通道也同理。通过上表的色彩对应的通道最大最小值，在查找颜色的时候便有了一个范围

```python
src = cv.imread('images/greenback.png')
hsv = cv.cvtColor(src, cv.COLOR_BGR2HSV)
plt.imshow(src)
```

```python
# hsv图像上找到绿色
mask = cv.inRange(hsv, (35, 43, 46), (77, 255, 255))
dst_not = cv.bitwise_not(mask)
dst_and = cv.bitwise_and(src, src, mask=dst_not )
plt.imshow(dst_and)


```
