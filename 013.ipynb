# 图像旋转
opencv中有三种图像的翻转方式：x轴翻转、Y轴翻转、xy轴同时翻转
- X轴翻转，flipcode = 0
- Y轴翻转, flipcode = 1
- XY轴翻转, flipcode = -1

```python
import cv2 as cv
import numpy as np
import utils
from matplotlib import pyplot as plt

src = cv.imread('images/flower.png')
utils.imshow(src)
```

```python
dst = cv.flip(src, 0)
utils.imshow(dst)
```

```python
dst = cv.flip(src, 1)
utils.imshow(dst)
```

```python
dst = cv.flip(src, -1)
utils.imshow(dst)




```
